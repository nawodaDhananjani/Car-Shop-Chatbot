```python
# Step 1: Install dependencies
!pip install transformers gradio torch

# Step 2: Import libraries
from transformers import pipeline
import gradio as gr

#  Step 3: Load a pre-trained model
chatbot = pipeline("text-generation", model="gpt2")  # Use GPT-2 (or swap with a better model)

#  Step 4: Define chatbot function
def carshop_chat(user_input):
    response = chatbot(user_input, max_length=100, num_return_sequences=1, temperature=0.7)
    return response[0]["generated_text"]

# Step 5: Create a Gradio Chatbot UI
chat_ui = gr.Interface(
    fn=carshop_chat,
    inputs=gr.Textbox(lines=2, placeholder="Ask me about car models, financing, services..."),
    outputs="text",
    title="🚗 Car Shop Chatbot",
    description="Ask me anything about car models, financing, or test drives!",
)

# Step 6: Launch chatbot
chat_ui.launch(share=True)
```
